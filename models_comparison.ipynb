```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_excel('models_comparison.xlsx')
df = pd.DataFrame(data)
```

```python
df
```

```python
price = []
quality = []
speed = []
names = []
mmlu = []

for index, row in df.iterrows():
    price.append(row['Ціна (USD/1m)'])
    quality.append(row['Artificial Analysis Quality Index'])
    speed.append(row['Швидкість (tk/s)'])
    names.append(row['Модель'])
    mmlu.append(row['Reasoning & Knowledge (MMLU) %'])
```

```python
lmsys_price = []
lmsys_quality = []
lmsys_names = []
lmsys = []

for index, row in df.iterrows():
    if not math.isnan(row['Communication (LMSys Chatbot Arena ELO Score)']):
        lmsys.append(row['Communication (LMSys Chatbot Arena ELO Score)'])
        lmsys_price.append(row['Ціна (USD/1m)'])
        lmsys_quality.append(row['Artificial Analysis Quality Index'])
        lmsys_names.append(row['Модель'])
```

```python
print(lmsys_names)
```

```python
# Создаем фигуру с тремя подграфиками
fig, axs = plt.subplots(1, 3, figsize=(15, 5))  # 1 строка, 3 столбца

# График 1: Цена / Качество
axs[0].scatter(price, quality, vmin=0, vmax=100)
axs[0].set_title("Ціна / Якість")
axs[0].set_xlabel("Ціна (USD/1m)")
axs[0].set_ylabel("Artificial Analysis Quality Index")
for i, name in enumerate(names):
    axs[0].annotate(name, (price[i], quality[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# График 2: Скорость / Качество
axs[1].scatter(speed, quality, vmin=0, vmax=100)
axs[1].set_title("Швидкість / Якість")
axs[1].set_xlabel("Швидкість (tk/s)")
axs[1].set_ylabel("Artificial Analysis Quality Index")
for i, name in enumerate(names):
    axs[1].annotate(name, (speed[i], quality[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# График 3: Цена / Скорость
axs[2].scatter(price, speed, vmin=0, vmax=100)
axs[2].set_title("Ціна / Шкидкість")
axs[2].set_xlabel("Ціна (USD/1m)")
axs[2].set_ylabel("Швидкість (tk/s)")
for i, name in enumerate(names):
    axs[2].annotate(name, (price[i], speed[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# Увеличиваем расстояние между графиками
plt.tight_layout()

# Показываем графики
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(15, 10))  # 2 строки, 2 столбца

# График 1: Ціна / MMLU
axs[0, 0].scatter(price, mmlu, vmin=0, vmax=100)
axs[0, 0].set_title("Ціна / MMLU")
axs[0, 0].set_xlabel("Ціна (USD/1m)")
axs[0, 0].set_ylabel("MMLU")
for i, name in enumerate(names):
    axs[0, 0].annotate(name, (price[i], mmlu[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# График 2: Якість / MMLU
axs[0, 1].scatter(quality, mmlu, vmin=0, vmax=100)
axs[0, 1].set_title("Якість / MMLU")
axs[0, 1].set_xlabel("Artificial Analysis Quality Index")
axs[0, 1].set_ylabel("MMLU")
for i, name in enumerate(names):
    axs[0, 1].annotate(name, (quality[i], mmlu[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# График 3: Ціна / LMSys
axs[1, 0].scatter(lmsys_price, lmsys, vmin=0, vmax=100)
axs[1, 0].set_title("Ціна / LMSys")
axs[1, 0].set_xlabel("Ціна (USD/1m)")
axs[1, 0].set_ylabel("LMSys")
for i, name in enumerate(lmsys_names):
    axs[1, 0].annotate(name, (lmsys_price[i], lmsys[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# График 4: Якість / LMSys
axs[1, 1].scatter(lmsys_quality, lmsys, vmin=0, vmax=100)
axs[1, 1].set_title("Якість / LMSys")
axs[1, 1].set_xlabel("Artificial Analysis Quality Index")
axs[1, 1].set_ylabel("LMSys")
for i, name in enumerate(lmsys_names):
    axs[1, 1].annotate(name, (lmsys_quality[i], lmsys[i]), fontsize=8, xytext=(5, 5), textcoords='offset points')

# Увеличиваем расстояние между графиками
plt.tight_layout()

# Показываем графики
plt.show()
```
